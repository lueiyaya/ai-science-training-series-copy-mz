````python
import requests

# 文件的URL
file_url = 'https://github.com/lueiyaya/ai-science-training-series-copy-mz/blob/main/07_AITestbeds/homework_files/Sambanova/BertLarge_4.out'

# 获取文件内容
response = requests.get(file_url)

# 确保请求成功
if response.status_code == 200:
    file_content = response.text
else:
    print("Failed to retrieve the file.")
    file_content = ""
````

````python
import re

# 定义一个正则表达式来查找包含average_loss的行
pattern = r"average_loss:(\d+\.\d+)"

# 在文件内容中搜索所有匹配项
matches = re.findall(pattern, file_content)

average_loss=[]
# 打印匹配到的所有average_loss值
for match in matches:
    average_loss.append(match)
````
